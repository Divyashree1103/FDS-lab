```python
import matplotlib.pyplot as plt
import numpy as np
year = [2014,2015,2016, 2017, 2018, 2019, 2020,2021]
placement_Computer = [80,82,85,90,92,94,96,100]
Plecement_civil = [78,80,84,88,90,89,89,88]
```

```python
#a)	Plot a line graph for above details.
plt.plot(year, placement_Computer, color ='black', 
         label ='placement_Computer')

plt.plot(year, Plecement_civil, color ='red', 
         label ='Placement_civil')
plt.xlabel('Years')
plt.ylabel('Placement')
plt.title('placements In Years')
  
plt.legend()
plt.show()
```

```python
#a)	Plot a Bar graph for above details.
plt.bar(year, placement_Computer, color ='black', 
         label ='placement_Computer',width = 0.4)

plt.bar(year, Plecement_civil, color ='red', 
         label ='Plecement_civil',width = 0.4)

plt.ylim(20, 140)

plt.xlabel('Years')
plt.ylabel('Placement')
plt.title('placements In Years')
  
plt.legend()
plt.show()
```

```python
#fill between.

import matplotlib.pyplot as plt

import numpy as np

x = np.arange(0.0, 2, 0.01)

y1 = np.sin(2* np.pi * x)

y2 = 0.8 * np.sin(4* np.pi *x)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, figsize=(6, 6))

ax1.fill_between (x, y1)

ax1.set_title('fill between y1 and 0')

ax2.fill_between(x, y1, 1)
ax2.set_title('fill between y1 and 1')

ax3.fill_between(x, y1, y2)
ax3.set_title('fill between y1 and y2')
ax3.set_xlabel('x')
fig.tight_layout()
```

```python
#Stack plot

import matplotlib.pyplot as plt

# List of Days
days = [1, 2, 3, 4, 5]

# No of Study Hours
Studying = [7, 8, 6, 11, 7]

#No of Playing Hours
playing = [8, 5, 7, 8, 13]

# Stackplot with X, Y, colors value 
plt.stackplot(days, Studying, playing, colors =['r', 'c'])

# Days
plt.xlabel('Days')

# No of hours
plt.ylabel('No of Hours')

# Title of Graph
plt.title('Representation of Study and \
Playing wrt to Days')

# Displaying Graph plt.show()
plt.show()
```

```python
#contour plot 
# Implementation of matplotlib function 
import matplotlib.pyplot as plt 
import numpy as np

feature_x = np.linspace (-5.0, 3.0, 70)
feature_y= np.linspace (-5.0, 3.0, 78)

# Creating 2-D grid of features
[X, Y] = np.meshgrid (feature_x, feature_y)

fig, ax = plt.subplots(1, 1)

Z = X ** 2 + Y ** 2

# plots filled contour plot
ax.contourf (X, Y, Z)

ax.set_title('Filled Contour Plot') 
ax.set_xlabel('feature_x') 
ax.set_ylabel('feature_y')

plt.show()
```
